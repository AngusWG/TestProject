```python
import rqdatac
rqdatac.init()
```

```python
df  = rqdatac.fund.all_instruments()
manager_df = rqdatac.fund.get_manager(df.order_book_id,expect_df=True)
manager_df = manager_df.reset_index()
```

```python
df_b = manager_df[(manager_df.end_date.isna()) |(manager_df.end_date>="20200101")]
```

```python
df_a = df_b[df_b['return']>0]
df_a = df_a.sort_index(by='return',ascending=False)
df_a
```

```python
name = df_a[:1000].name.value_counts()[:10].index.to_list()
print(name)
df_c = df_a[df_a.name.isin(name)]
df_c['symbol'] = [i.symbol for i in rqdatac.fund.instruments(df_c.order_book_id)]
a = df_c[:15]
a
```

```python
liu_df = manager_df[manager_df.name=="刘格菘"].sort_index(by='return',ascending=False)
liu_nav_df = rqdatac.fund.get_nav(liu_df.order_book_id,start_date=20191201, expect_df=True)
liu_nav_df.reset_index(inplace=True)  
liu_nav_df.groupby("order_book_id").mean().sort_index(by='change_rate',ascending=False)
```
